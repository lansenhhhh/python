## 跟踪点的轨迹

## 合成图像

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation 
fig,ax=plt.subplots(figsize=(5,4))
x=np.arange(0,2*np.pi,0.01)
y=np.sin(x)
def update(i):
    ax.plot(x[i],y[i],'b.',lw=0.01)
    label='frame={0}'.format(i)
    ax.set_xlabel(label)
    return ax 
anim=FuncAnimation(fig,update,frames=np.arange(0,len(x)),interval=10)
anim.save('F:python/data/sin2_anim.gif',fps=50)
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation 
fig,ax=plt.subplots(figsize=(5,4))
x=np.arange(0,2*np.pi,0.01)
y=np.sin(3*x)
ax.axis([0,2*np.pi,-1,1])
line,=ax.plot(x,y)
def update(i):
    label='frame{0}'.format(i)
    ax.set_xlabel(label)
    line.set_xdata(x);
    line.set_ydata(np.sin(3*x+i*np.ones([1,len(x)])))
    return line,ax
anim=FuncAnimation(fig,update,frames=np.arange(0,2*np.pi,0.1),interval=20)
plt.show()
```
